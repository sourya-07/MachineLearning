Data Manipulation and Analysis with Pandas

It's a key tasks in any data science or data analysis project. Pandas provides a wide range of functions for ata manipulation and analysis, making it easier to clean, trasfrom, and exact insights from tdata. Now, I will cover varios data manipulation and analysis techniques using Pandas.

```python
import pandas as pd
```

```python
# Handling Missing Values
df.isnull().any(axis=1)
```

```python
df.isnull().sum()
```

```python
df_filled = df.fillna(0)
```

```python
# Filling missing values with the mean of the column
df['Calories_fillNA'] = df['Calories'].fillna(df['Calories'].mean())
df
```

```python
df.dtypes
```

```python
# Renaming Columns
df = df.rename(columns={'Date': 'workout Date'})
df
```

```python
# Change datatypes
df['Maxpulse_new'] = df["Maxpulse"].astype(float)
df.head()
```

```python
# Data Aggregating and Grouping
df.head()
```

```python
grouped_mean = df.groupby('Pulse')['Calories'].mean()
print(grouped_mean)
```

```python
# Aggretage multiple functions
grouped_agg = df.groupby('Duration')['Calories'].agg(['mean','sum', 'count'])
grouped_agg
```

```python
# Merging and joining Dataframes
# create sample Dataframes
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Values1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
# Merge Dataframe on the 'Key Columns'
pd.merge(df1, df2, on='Key',how='inner')
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```

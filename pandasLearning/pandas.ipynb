```python
# A Pandas series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd

data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
# Create a series from dictionary
data = {"a": 1, "b": 2, "c": 3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10, 20, 30]
index = ['a', 'b', 'c']
pd.Series(data, index=index)
```

```python
# Dataframe
# Create a Dataframe from a dic of list

data = {
    'Name': ['Krish', 'John', 'Jack'],
    "Age": [25, 30, 45],
    "City": ["Banglore", 'New York', 'Florida']   
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# Create a dataframe from a list of dictionaries:-

data = [
    {'Name': 'Krish', "Age": 20, "City": "Banglore"},
    {'Name': 'Sourya', "Age": 18, "City": "Banglore"},
    {'Name': 'Keshav', "Age": 20, "City": "Banglore"},
    {'Name': 'Dipu', "Age": 21, "City": "Banglore"},   
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
